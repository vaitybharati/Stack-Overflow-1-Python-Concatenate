# Stack Overflow-1 Python Concatenate

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Time':['01:00:00','01:00:00','01:00:01','01:00:01'], 
                   'DecimalSecond':['14','178','158','75']}) 
```

```python
df['Time2'] = df[['Time','DecimalSecond']].apply(lambda x: ' '.join(x), axis = 1)
print(df)
```
